```python
import requests

def fetch_fmp_metadata(ticker):
    url = f"https://financialmodelingprep.com/api/v3/profile/{ticker}?apikey=uoZUZtpfrOdmQdWdpwArSVOBH9wI7ob4"
    response = requests.get(url)
    data = response.json()

    # If API returns error object instead of list
    if isinstance(data, dict):
        print(f"❌ FMP error for {ticker}: {data}")
        return None
    
    if not data:  # empty list
        print(f"❌ No metadata found for {ticker}")
        return None
    
    d = data[0]  # safe now
    
    return {
        "ticker": ticker,
        "company_name": d.get("companyName"),
        "sector": d.get("sector"),
        "industry": d.get("industry"),
        "exchange": d.get("exchange"),
        "country": d.get("country"),
        "market_cap": d.get("marketCap")
    }
```

```python
records = []

for t in TICKERS:
    print("Fetching:", t)
    rec = fetch_fmp_metadata(t)
    if rec:
        records.append(rec)

metadata_df = pd.DataFrame(records)
metadata_df
```
